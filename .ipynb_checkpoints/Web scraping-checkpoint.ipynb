```python
import urllib3
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
warnings.filterwarnings('ignore')
```

```python
url='https://www.yelp.com/search?find_desc=Sandwiches&find_loc=Boston%2C+MA+2108&ns=1'
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
```

```python
response.data
```

```python
soup=BeautifulSoup(response.data, "lxml")
print(soup.prettify())
```

```python
attributes=soup.find_all("div", class_="secondary-attributes")
for attr in attributes:
        print attr.address.get_text()
```

```python
#recognises only the first div inside the div for biz-listing-large. does not find secondary-attributes
main_attr=soup.find_all("div",class_="biz-listing-large")
for ma in main_attr:
    if ma.div["class"]==["main-attributes"]:
        if ma.div.span["class"]==["indexed-biz-name"]:
            print "Name:"+ma.div.span.get_text()
        elif ma.div.span["class"]==["review-count rating-qualifier"]:
            print "Review:"+ma.div.span.get_text()
    elif ma.div["class"]==["secondary-attributes"]:
        if ma.div.span["class"]==["neighborhood-str-list"]:
            print "Location:"+ma.div.span.get_text()
```

```python
sec=soup.find_all("div",class_="search-result natural-search-result")
for s in sec:
    if s.div["class"]==["biz-listing-large"]:
            if s.div.div["class"]==["neighborhood-str-list"]:
                print s.div.span
                print "\n\n\n"
            else:
                print s.div.div
```
